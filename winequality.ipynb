```python
plt.figure(figsize=(10,11))
sns.heatmap(df.corr(),annot=True,cmap="YlGnBu")
plt.plot()
```

```python
labels=["White","red"]
df['type'].value_counts().plot(kind='pie', title='Types of wine',labels=labels,figsize=(5,11))
plt.show()
```

```python
plt.figure(figsize=(12,9))
plt.title("Influence of Alcohol in wine quantity")
sns.scatterplot(df["alcohol"],df["quality"],s=30,hue=df["type"])
plt.show()
```
